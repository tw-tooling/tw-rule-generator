```python

def generate_rule(rule_name: str = "test", base_tile: int = 1, border_tile: int = 16, corner_tile_outer: int = 32, corner_tile_inner: int=48) -> str:
    top_tile = f"{border_tile}"
    left_tile = f"{border_tile} ROTATE YFLIP"
    right_tile = f"{border_tile} ROTATE"
    bottom_tile = f"{border_tile} YFLIP"
    top_left_tile_outer = f"{corner_tile_outer}"
    top_right_tile_outer = f"{corner_tile_outer} ROTATE"
    bottom_left_tile_outer = f"{corner_tile_outer} XFLIP YFLIP ROTATE"
    bottom_right_tile_outer = f"{corner_tile_outer} XFLIP YFLIP"
    top_left_tile_inner = f"{corner_tile_inner}"
    top_right_tile_inner = f"{corner_tile_inner} ROTATE"
    bottom_left_tile_inner = f"{corner_tile_inner} XFLIP YFLIP ROTATE"
    bottom_right_tile_inner = f"{corner_tile_inner} XFLIP YFLIP"

    rule = f'''
[{rule_name}]
Index {base_tile}
BaseTile

#top
Index {top_tile}
Pos 0 -1 EMPTY
Pos 0 1 FULL

#left
Index {left_tile}
Pos -1 0 EMPTY
Pos 1 0 FULL

#right
Index {right_tile}
Pos -1 0 FULL
Pos 1 0 EMPTY

#bottom
Index {bottom_tile}
Pos 0 -1 FULL
Pos 0 1 EMPTY

#top_left_outer
Index {top_left_tile_outer}
Pos 0 -1 EMPTY
Pos -1 0 EMPTY
Pos 1 0 FULL
Pos 0 1 FULL

#top_right_outer
Index {top_right_tile_outer}
Pos 0 -1 EMPTY
Pos -1 0 FULL
Pos 1 0 EMPTY
Pos 0 1 FULL

#bottom_left_outer
Index {bottom_left_tile_outer}
Pos 0 -1 FULL
Pos -1 0 EMPTY
Pos 1 0 FULL
Pos 0 1 EMPTY

#bottom_right_outer
Index {bottom_right_tile_outer}
Pos 0 -1 FULL
Pos -1 0 FULL
Pos 1 0 EMPTY
Pos 0 1 EMPTY

#top_left_inner
Index {top_left_tile_inner}
Pos -1 -1 EMPTY
Pos 1 -1 FULL
Pos -1 1 FULL
Pos 1 1 FULL
Pos 0 -1 FULL
Pos -1 0 FULL
Pos 0 1 FULL
Pos 1 0 FULL

#top_right_inner
Index {top_right_tile_inner}
Pos -1 -1 FULL
Pos 1 -1 EMPTY
Pos -1 1 FULL
Pos 1 1 FULL
Pos 0 -1 FULL
Pos -1 0 FULL
Pos 0 1 FULL
Pos 1 0 FULL

#bottom_left_inner
Index {bottom_left_tile_inner}
Pos -1 -1 FULL
Pos 1 -1 FULL
Pos -1 1 EMPTY
Pos 1 1 FULL
Pos 0 -1 FULL
Pos -1 0 FULL
Pos 0 1 FULL
Pos 1 0 FULL

#bottom_right_inner
Index {bottom_right_tile_inner}
Pos -1 -1 FULL
Pos 1 -1 FULL
Pos -1 1 FULL
Pos 1 1 EMPTY
Pos 0 -1 FULL
Pos -1 0 FULL
Pos 0 1 FULL
Pos 1 0 FULL

    '''
    return rule.strip()
```

```python
import gradio as gr

simple_mapper_interface = gr.Interface(fn=generate_rule, inputs=["text","number","number","number","number"], outputs="code", allow_flagging='never')
empty_interface = gr.Interface(lambda nothing: nothing, inputs=["text"], outputs=["code"], allow_flagging='never')
tab_interface = gr.TabbedInterface([simple_mapper_interface, empty_interface], ['Teeworlds Rule Generator', 'TODO'])
tab_interface.launch()
```
